```python
score_NAR = score_NAR[score_NAR['User'] != "User_1"]
mapping = {'User_2': "User_1", 'User_3': "User_2", 'User_4': "User_3", 'User_5': "User_4", 'User_6': "User_5", 'User_7': "User_6", 'User_8': "User_7", 'User_9': "User_8", 'User_10': "User_9",'User_11': "User_10"}
score_NAR['User'] = score_NAR['User'].apply(lambda x: mapping[x])

score_AR = score_AR[score_AR['User'] != "User_1"]
# mapping = {'User_2': "User_1", 'User_3': "User_2", 'User_4': "User_3", 'User_5': "User_4", 'User_6': "User_5", 'User_7': "User_6", 'User_8': "User_7", 'User_9': "User_8", 'User_10': "User_9",'User_11': "User_10"}
score_AR['User'] = score_AR['User'].apply(lambda x: mapping[x])
```

```python
aggregation_functions = {'User': 'first', 'Section': 'first','Jump':'max', 'Positive':'max','Negative':'max', }
score_NAR = score_NAR.groupby(score_NAR['timestamp']).aggregate(aggregation_functions).reset_index()


score_AR = score_AR.groupby(score_AR['timestamp']).aggregate(aggregation_functions).reset_index()

```

```python
rt = score_NAR.groupby(['User','Section']).apply(
    lambda x: (
        ( (x['timestamp'] - x['timestamp'].min()).apply(lambda x: x.seconds) ) / 
        ( (x['timestamp'].max() - x['timestamp'].min()).seconds )
    )
)
score_NAR['n_ts'] = score_NAR.apply(lambda x: rt.loc[ x['User'],x['Section'] ][x.name], axis=1)


rt = score_AR.groupby(['User','Section']).apply(
    lambda x: (
        ( (x['timestamp'] - x['timestamp'].min()).apply(lambda x: x.seconds) ) / 
        ( (x['timestamp'].max() - x['timestamp'].min()).seconds )
    )
)
score_AR['n_ts'] = score_AR.apply(lambda x: rt.loc[ x['User'] ,x['Section']][x.name], axis=1)
```

```python
score_NAR['Jump'] = score_NAR['Jump'].astype(float)
score_NAR['Positive'] = score_NAR['Positive'].astype(float)
score_NAR['Negative'] = score_NAR['Negative'].astype(float)
score_NAR['n_ts'] = score_NAR['n_ts'].astype(float)

score_AR['Jump'] = score_AR['Jump'].astype(float)
score_AR['Positive'] = score_AR['Positive'].astype(float)
score_AR['Negative'] = score_AR['Negative'].astype(float)
score_AR['n_ts'] = score_AR['n_ts'].astype(float)
```
